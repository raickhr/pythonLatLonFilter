```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 1024)[:-1]
y  = np.sin(x) 
y1 = 0.1 *np.sin(50*x)
y1[0:512-25] = 0
y1[512+25:-1] = 0

y = y + y1
#yorig = np.sin(x)
```

```python
#plt.plot(x,yorig, label = 'orig')
plt.plot(x,y, label='both')
plt.legend()
```

```python
klen = int(len(x)/49)
kernel = np.ones(klen)
kernel = kernel/np.sum(kernel)
ylow = scipy.ndimage.convolve1d(y, kernel, axis=-1, mode='wrap')
```

```python
plt.plot(x,ylow)
```

```python
yhigh = y - ylow
plt.plot(x,yhigh)
```
